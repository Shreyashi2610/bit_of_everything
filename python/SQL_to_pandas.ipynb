#### 1.<br>select * <br>from country<br>where population > 50000000<br>order by population DESC limit 10
#### Description
Printing the top 10 most populated countries which have population more than 50 million 

```python
country[country['Population']>50000000].sort_values('Population',ascending=False).head(10)
```

#### 2.<br>select Continent, count(*) As Number_Countries, sum(population) As Population<br>from country<br>where population > 0<br>group by Continent<br>order by 1 ASC
#### Description
Printing Continents with their number of countries and total population, in alphabetical order

```python
country[country['Population']>0].groupby('Continent').agg({"Name": "count",
                                                           "Population": "sum"}).sort_values('Continent')
```

#### 3.<br>select city.Name As City, city.population<br>from city<br>inner join country ON city.CountryCode = country.code<br>where country.code = 'USA'<br>order by city.population DESC limit 10
#### Description
Printing all the cities in USA with their populations

```python
usa_city = pd.merge(city[['CountryCode','Name','Population']],
                    country[['Code','Name']][country['Code'] == 'USA'],
                    left_on = 'CountryCode',
                    right_on = 'Code',
                    how = 'inner')[['Name_x','Population']].sort_values('Population',ascending=False).head(10)
usa_city.rename(columns = {'Name':'City'}, inplace = True)
usa_city
```

#### 4.<br>select country.Name, Language, (Percentage * population) / 100<br>from countrylanguage<br>inner join country on countrylanguage.CountryCode = country.code<br>where IsOfficial = True<br>order by 3 DESC limit 10
#### Description
Printing the top 10 official language of countries in terms of the population speaking that language

```python
off_language = pd.merge(countrylanguage[countrylanguage['IsOfficial']=='T'][['Language','CountryCode','Percentage']],
                    country[['Name','Population','Code']],
                    left_on = 'CountryCode',
                    right_on = 'Code',
                    how = 'inner')[['Name','Language','Percentage','Population']]
off_language['(Percentage * population) / 100'] = off_language['Percentage']*off_language['Population']
off_language[['Name',
              'Language',
              '(Percentage * population) / 100']].sort_values('(Percentage * population) / 100',
                                                              ascending=False).head(10)
```

#### 5.<br>select Language, sum((Percentage * population) / 100)<br>from countrylanguage<br>inner join country on countrylanguage.CountryCode = country.code<br>group by Language<br>order by 2 desc limit 5
#### Description
Printing the top 5 official languages according the total population that speaks them

```python
off_language = pd.merge(countrylanguage,
                    country[['Population','Code']],
                    left_on = 'CountryCode',
                    right_on = 'Code',
                    how = 'inner')[['Language','Percentage','Population']]

off_language['(Percentage * population) / 100'] = off_language['Percentage']*off_language['Population']

off_language.groupby('Language').agg({"(Percentage * population) / 100": "sum"}).sort_values('(Percentage * population) / 100',
                                                              ascending=False).head(5)
```
